# Filtering Columns & Rows in Pandas

```python
# import pandas

import pandas as pd
```

```python
# read csv

df = pd.read_csv(r"C:\Users\shass\my_github\my_practice_repository\pandas_for_beginners\datasets\world_population.csv")
```

```python
# view dataframe

df
```

## Filtering Data in Columns

```python
# filtering data in the 'Rank' column where the value is <= '10'

df[df['Rank'] <= 10]
```

## 'Isin' Filter

```python
# searching our dataframe for specific countries based on a list

# creating the variable which is a list called 'specific_countries' & naming the countries we want to include in that list

specific_countries = ['Bangladesh', 'Brazil']

# filtering the dataframe to search for the specific country list using the 'isin' function

df[df['Country'].isin(specific_countries)]
```

## 'Contains' Filter

```python
# filtering the dataframe to search for the countries with 'United' in the name using the 'contains' function

df[df['Country'].str.contains('United')]
```

## Using Indexing

```python
# creating a seperate dataframe to work with & setting the index as the country

df2 = df.set_index('Country')
df2
```

```python
# using a filter to specify which columns we wish to view

df2.filter(items = ['Continent', 'CCA3'])
```

## Filtering by the Columns Axis (axis = 1)

```python
# we can specify which axis we want to search & filter these values, the 'rows' axis = 0, and the 'columns' axis = 1

df2.filter(items = ['Continent', 'CCA3'], axis = 1)
```

## Filtering by the Rows Axis (axis = 0)

```python
# we can specify which axis we want to search & filter these values, the 'rows' axis = 0, and the 'columns' axis = 1

df2.filter(items = ['Zimbabwe', 'Zambia'], axis = 0)
```

```python
# similar to how we filtered for 'United' previously we can use the 'like' search to look for values in the rows which meet
# the filter criteria

df2.filter(like = 'United', axis = 0)
```

## 'Loc' Filter

```python
# The .loc[] attribute is used to access a group of rows and columns by labels or a boolean array
# So, df2.loc['United States'] means that we are selecting the row(s) with the label 'United States' from the DataFrame df2.
# This will return a Series object if there is only one row with the label 'United States', or a DataFrame if there are 
# multiple rows with this label


df2.loc['United States']
```

## 'Iloc' Filter

```python
# The .iloc[] attribute is used to access a group of rows and columns by integer position(s)
# .iloc[] is an integer-location based indexing for selection by position. Unlike .loc[], which uses labels, .iloc[] uses 
# integer positions to make selections
# So, df2.iloc[3] means that we are selecting the fourth row of the DataFrame df2 based on its integer position

df2.iloc[3]
```

## 'Order By' Filter

```python
# filters the rows of the DataFrame df to include only those with a 'Rank' value less than 10, and then sorts those rows 
# first by the 'Continent' column in descending order and then by the 'Country' column in ascending order. The result is a 
# new DataFrame containing the filtered and sorted data.

df[df['Rank'] < 10].sort_values(by = ['Continent', 'Country',], ascending = [False, True])
```
