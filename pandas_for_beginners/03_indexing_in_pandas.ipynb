# Indexing in Pandas

#### In Pandas, indexing refers to the process of selecting specific rows and columns from a DataFrame or Series. It allows you to efficiently access, modify, and organize your data

```python
import pandas as pd
```

```python
# reading in csv & setting the index to 'Country', in the results you'll see the index is in on the left, bold & lower than others

df = pd.read_csv(r"C:\Users\shass\my_github\my_practice_repository\pandas_for_beginners\datasets\world_population.csv", index_col = 'Country')

df
```

```python
# reseting the dataframe index

df.reset_index(inplace=True)

df
```

```python
# Other ways to setting the index, in this example we set the index manually, we also need to include the 'inplace' function
# so that the results save to the dataframe & can be used going forward

df.set_index('Country', inplace = True)

df
```

## 'Loc' function

```python
# the loc function selects rows using row labels (e.g. tea )

df.loc['Albania']
```

```python
# the iloc function selects rows using their integer positions (staring from 0 and going up by one for each row)

df.iloc[1]
```

## Multi-Indexing

#### A MultiIndex, also known as a multi-level index or hierarchical index, allows you to have multiple columns acting as a row identifier, while having each index column related to another through a parent/child relationship

```python
# reseting the index again

df.reset_index(inplace=True)

df
```

```python
# setting a multi-index of both the 'Continent' & 'Country'

df.set_index(['Continent', 'Country'], inplace = True)

df
```

```python
# sorting the multi-index so by 'Continent' & 'Country'

df.sort_index()
```

```python
# we can use the loc function to return the specific rows we want based on the index
# as we have multi-indexed the first column it will search in the first column

df.loc['Africa']
```

```python
# if we want to search in the second column like before we need to include two search paramaters

df.loc['Africa', 'Angola']
```
