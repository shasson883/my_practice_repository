# Group by and Aggregating

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\shass\my_github\my_practice_repository\pandas_for_beginners\datasets\flavors.csv")

df
```

## Shortcut Function 'Describe' to Quickly View Data

```python
df.groupby('Base Flavor').describe()
```

## Group By Function

```python
# grouping the df by 'base flavor' & assigning it to the variable 'group_by_frame'

group_by_frame = df.groupby('Base Flavor')
```

```python
# calling the variable 'group_by_frame' & returning the 'mean' of the numerical (ints) values in the table

group_by_frame.mean()
```

```python
# another method of compiling this code

df.groupby('Base Flavor').mean()
```

```python
# same code but now we're looking at the 'count' instead of 'mean'

df.groupby('Base Flavor').count()
```

```python
# same code but now we're looking at the 'count'

df.groupby('Base Flavor').count()
```

```python
# same code but now we're looking at the 'min', it also includes the string values when bringing in the min, it looks at the
# index of the first letter of the string value & returns the min, e.g. 'No'. Note that for Chocolate the values are all 
# equal to 'Yes' so this is the min value

df.groupby('Base Flavor').min()
```

```python
# same code but now we're looking at the 'max', it also includes the string values when bringing in the max

df.groupby('Base Flavor').max()
```

```python
# same code but now we're looking at the 'sum'

df.groupby('Base Flavor').sum()
```

## Aggregate Function

```python
# The agg function calculates multiple aggregation functions (mean, max, count, sum) on the 'Flavor Rating' column 
# within each group of 'Base Flavor'. We create a dictionary based on the column 'Flavor Rating', then return the different
# numerical outputs based on the different calculations

df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```

```python
# We can include additional aggregation functions within this code

df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'], 'Texture Rating': ['mean', 'max', 'count', 'sum']})
```

## Grouping by Multiple Columns

```python
# we can add additional columns to our grouping function by adding them in as a list

df.groupby(['Base Flavor', 'Liked']).mean()
```

```python
# we can add the aggergation function to the groupby to perform the same calculations on seperate grouped values

df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```
