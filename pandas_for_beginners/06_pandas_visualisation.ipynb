# Pandas Visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\shass\my_github\my_practice_repository\pandas_for_beginners\datasets\ice_cream_ratings.csv", index_col = 'Date')
df
```

## Line Chart Plot

```python
# creating a generic plot, by default without specifying the details it'll create a line plot

df.plot()
```

```python
# we can also specify that it's a line plot

df.plot(kind = 'line')
```

```python
# we can split out the plots into subplots

df.plot(kind = 'line', subplots = True)
```

```python
# we can customise the title & axis labels using the following

df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

## Bar Chart Plot

```python
# creating a standard bar chart

df.plot(kind = 'bar')
```

```python
# creating a stacked bar chart, note you can also use the same title & label code for adjustments

df.plot(kind = 'bar', stacked = True, title = 'Ice Cream Ratings', ylabel = 'Scores', xlabel = 'Date')
```

```python
# specifying what data to include from the dataframe in the plot

df['Flavor Rating'].plot(kind = 'bar', stacked = True, ylabel = 'Scores', title = 'Flavor Ratings')
```

```python
# generates a horizontal stacked bar plot using data from the DataFrame df. Each row in the DataFrame corresponds to a bar, 
# and the bars are stacked to represent multiple categories within each row. It's useful for visualizing the composition of 
# each row's values in a horizontal format.

df.plot.barh(stacked = True)
```

## Scatter Plot

```python
# creates a scatter plot using data from the DataFrame df. It places points based on 'Texture Rating' and 'Overall Rating' 
# columns. The s parameter sets point size to 500 for better visibility. It's a visual representation of the relationship 
# between these two variables.

df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, title = 'Texture Rating vs Overall Rating Scatter Chart')
```

## Histogram Plot

```python
# generates a histogram plot using data from the DataFrame df. It divides the data into bins and displays the frequency 
# distribution of values. The bins parameter sets the number of bins to 50, resulting in a visual representation of the 
# data's distribution.

# histograms are really good for showing distributions of variables

df.plot.hist(bins = 50)
```

## Box Plot

```python
# creates a box plot using data from the DataFrame df. The box plot visually displays the distribution of data's quartiles, 
# showing median, interquartile range, and potential outliers. It helps visualize the spread and central tendency of the data.

df.boxplot()
```

## Area Plot

```python
# generates an area plot using data from the DataFrame df. It displays the data as stacked areas, showing how the values of 
# different columns change over time or another continuous variable. It's useful for visualizing the cumulative contribution 
# of each column's values.

df.plot.area()
```

## Pie Chart Plot

```python
# creates a pie chart using data from the DataFrame df. It visualizes the distribution of values in the 'Flavor Rating' column. 
# The figsize parameter sets the size of the chart. It's suitable for showing the relative proportions of different flavor 
# ratings.

df.plot.pie(y = 'Flavor Rating', figsize = (10, 10))
```

### Adjusting the Size of the Chart

```python
# you can adjust the size of the chart using this code

df.plot.area(figsize = (10,5))
```
