# Data Cleaning in Pandas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\shass\my_github\my_practice_repository\pandas_for_beginners\datasets\customer_call_list.xlsx")
df
```

## Remove Duplicates

```python
# drop duplicates & save results as new dataframe

df = df.drop_duplicates()
df
```

## Removing Unwanted Columns

```python
# drop unwanted columns (you could also use inplace = True within the brackets)

df = df.drop(columns = 'Not_Useful_Column')
df
```

## Removing Unwanted Characters

```python
# lstrip removes values off the left had side of the string based on what you specify in the code
# to ensure that changes are applied to the specific column, ensure that you specify that column when assigning the arguement
# to the df

df['Last_Name'] = df['Last_Name'].str.lstrip('...')
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df['Last_Name'] = df['Last_Name'].str.rstrip('_')

df
```

```python
# can also write the code like this, though it's a little more confusing
# In this code, the 'Last_Name' column in the DataFrame df has its trailing characters removed. Specifically, any trailing 
# occurrences of '1', '2', '3', '.', '_', or '/' are stripped from the values in the 'Last_Name' column.

df['Last_Name'] = df['Last_Name'].str.rstrip('123._/')
```

# Replacing Unwanted Characters

```python
# In this code, the 'Phone_Number' column in the DataFrame df undergoes a string replacement. All non-alphanumeric characters 
# (symbols, spaces, etc.) are replaced with an empty string, effectively removing them from the phone number values.

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '')
df
```

```python
# converts all values in the 'Phone_Number' column of DataFrame df to strings. It uses a lambda function with the apply method 
# to ensure that each value is treated as a string, even if it was originally of a different data type.
# this is essential for the next step where we'll be formatting the output

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
# This code applies a lambda function to the 'Phone_Number' column in DataFrame df. It restructures each phone number by 
# adding hyphens to format it as "xxx-xxx-xxxx", where 'x' represents individual digits. This effectively changes the format 
# of the phone numbers in the column.

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# removing those values that we don't want in the 'Phone_Number' column

df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

## Splitting Columns

```python
# the 'Address' column in DataFrame df is split into three separate columns: 'Street_Address', 'State', and 'Zip_Code'. The 
# str.split method is used with , as the separator, and the expand=True argument ensures that the split components are assigned 
# to separate columns. This effectively separates the address information into its individual components.

df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', 2, expand = True)
df
```

## Using Replace to Update Values to be Uniform

```python
# updating the values in two of the columns so that they are uniform

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')

df
```

## Replacing all Missing Values with Value

```python
# fills all missing (NaN) values in the DataFrame df with empty strings (''). It replaces any NaN values across all columns 
# with empty strings, effectively removing the missing data by filling it with blanks.

df = df.fillna('')

# some additional cleaning
df['Paying Customer'] = df['Paying Customer'].str.replace('N/A', '')

df
```

## Removing Unwanted Rows

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
        
df
```

```python
# This code iterates through the rows of DataFrame df. For each row, it checks if the value in the 'Do_Not_Contact' column is 
# 'Y'. If the condition is met, it removes (drops) the entire row using the drop function with inplace=True, meaning the 
# changes are applied directly to the DataFrame. After the loop, the modified DataFrame df is displayed. Essentially, this 
# code filters out rows where the 'Do_Not_Contact' column has the value 'Y', resulting in a DataFrame without those rows.

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
        
df
```

```python
#This code loops through the rows of DataFrame df. For each row, it checks if the 'Phone_Number' column value is empty (''). 
# If the condition is true, it drops the entire row using the drop function with inplace=True. After the loop, the modified 
# DataFrame df is displayed. This code effectively filters out rows where the 'Phone_Number' column is empty, resulting in 
# a DataFrame without those rows.

for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
        
df
```

```python
# the above line of code can also be written as follows

df = df.dropna(subset = 'Phone_Number'), inplace = True)
```

## Resetting the Index

```python
# This code resets the index of DataFrame df while dropping the previous index. It reassigns a new sequential index to the 
# DataFrame's rows, eliminating the previous index values.

df = df.reset_index(drop = True)
df
```
