# Exploratory Data Analysis (EDA) in Python

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Setting Floats Decimal Places to 2

```python
# This code sets a display option in Pandas. It formats floating-point numbers to have two decimal places when shown in 
# DataFrames. It affects how numeric values are presented for better readability.

pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = pd.read_csv(r"C:\Users\shass\my_github\my_practice_repository\pandas_for_beginners\datasets\world_population.csv")
df
```

### Quickly Displaying Info About Your Data

```python
# displays concise information about DataFrame df. It includes the data types of columns, non-null counts, and memory usage. 
# It's useful for quickly understanding the structure and memory usage of the DataFrame.

df.info()
```

```python
# This code generates summary statistics of the numerical columns in DataFrame df. It provides information like mean, standard 
# deviation, minimum, maximum, and quartiles, helping to understand the distribution and characteristics of the data.

df.describe()
```

### Understanding Missing Data

```python
# counts the number of missing values (NaN) in each column of DataFrame df. It returns a series where each value represents the 
# count of missing values in the corresponding column. This is useful for identifying columns with incomplete data.

df.isnull().sum()
```

### Understanding Unique Values

```python
# calculates the number of unique values in each column of DataFrame df. It returns a series with the count of distinct values 
# in each column, providing insights into the diversity of data within each column.

df.nunique()
```

### Sorting & Exploring

```python
# sorts DataFrame df in descending order based on the values in the '2022 Population' column and displays the top 10 rows with 
# the highest population values. It allows you to quickly identify the most populous entries in the dataset.

df.sort_values(by = '2022 Population', ascending = False).head(10)
```

### Looking at Correlations

```python
# calculates the correlation matrix for the numerical columns in DataFrame df. It shows the pairwise correlation coefficients 
# between columns, helping to understand how variables are related and potentially indicating patterns or dependencies in the data

df.corr()
```

```python
# generates a heatmap using the Seaborn library, illustrating the correlation matrix of DataFrame df. The annot=True 
# parameter adds numeric values to the heatmap cells, aiding in understanding the correlation between different numerical 
# variables in the dataset.

sns.heatmap(df.corr(), annot = True)

# adjusts the default figure size for Matplotlib plots. It sets the width to 10 inches and the height to 7 inches, allowing for 
# larger and more visually appealing plots when using Matplotlib for visualization.

plt.rcParams['figure.figsize'] = (10, 7)

# displays any pending Matplotlib plots. It's necessary to include this command to actually show the plots you've created 
# using Matplotlib.

plt.show()
```

### Grouping Values

```python
# DataFrame df is grouped by the 'Continent' column, and then the mean values for each numerical column within each continent 
# group are calculated. The result is sorted in descending order based on the '2022 Population' column. The sorted and 
# aggregated data is assigned to a new DataFrame named df2.

df2 = df.groupby('Continent').mean().sort_values(by = '2022 Population', ascending = False)
df2
```

```python
# filters DataFrame df to show only rows where the 'Continent' column contains the string 'Oceania'. It displays a subset of 
# the original DataFrame containing data related to the continent of Oceania.

df[df['Continent'].str.contains('Oceania')]
```

```python
# groups DataFrame df by the 'Continent' column, calculates the mean values for each numerical column within each continent 
# group, and then sorts the result in descending order based on the '2022 Population' column. It provides a summary of average 
# values for each continent, ordered by population.

df.groupby('Continent').mean().sort_values(by = '2022 Population', ascending = False)
```

#### Use Case - Which country has grown the fastest?

```python
# if we want to see all of the columns in a dataset

df.columns
```

```python
# can update our df2 so that we're only pulling through the columns we want, i.e. the yearly columns
# selecting the columns in this way will ensure that they are in the correct order when we go to look at the line graph

df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by = '2022 Population', ascending = False)
df2
```

```python
# transpose df2 dataframe so that the column headers are now the index rows

df3 = df2.transpose()
df3
```

```python
# plotting using a line plot, we can clearly see that Asia is the fastest growing continent in the data

df3.plot(figsize = (15, 10))
```

## Looking at Outliers Using Boxplots

```python
# boxplots can be really good at showing outliers in comparison to other values

df.boxplot(figsize = (20, 10))
```

## Looking at Specific DataTypes

```python
# selects columns from DataFrame df that have numerical data types. It filters and returns a new DataFrame containing only the 
# columns that contain numerical values, excluding columns with non-numeric data types.

df.select_dtypes(include = 'number')
```

```python
# selects columns from DataFrame df that have object (string) data types. It filters and returns a new DataFrame containing 
# only the columns that contain string or text-based values, excluding columns with numerical data types.

df.select_dtypes(include = 'object')
```
