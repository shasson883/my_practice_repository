# If - Elif - Else Statements

```python
# if statement

if 25 > 10:
    print('It worked!')
```

```python
# if - else statements

if 25 < 10:
    print('It worked!')
else:
    print('It did not work...')
```

```python
# if - else - elif statements

if 25 < 10:
    print('It worked!')
elif 25 < 20:
    print('elif worked!')
elif 25 < 30:
    print('elif #2 worked!')  
else:
    print('It did not work...')
```

```python
# if - else - elif and or statements

if (25 < 10) or (1 < 3):
    print('It worked!')
elif 25 < 20:
    print('elif worked!')
elif 25 < 30:
    print('elif #2 worked!')  
else:
    print('It did not work...')
```

```python
# nested if statement

if (25 < 10) or (1 < 3):
    print('It worked!')
    if 10 > 5:
        print('This nested if statement worked!')
elif 25 < 20:
    print('elif worked!')
elif 25 < 30:
    print('elif #2 worked!')  
else:
    print('It did not work...')
```
