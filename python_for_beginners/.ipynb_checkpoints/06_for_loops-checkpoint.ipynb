# For Loops

```python
# This line of code is creating a list called integers, which contains five elements, the integers from 1 to 5

integers = [1,2,3,4,5]
```

```python
# This line is the start of a for loop. A for loop in Python is used for iterating over a sequence 
# (that is either a list, a tuple, a dictionary, a set, or a string). In this case, the sequence is 
# the integers list. The variable number will take on the value of each element in the integers list 
# one by one in each iteration of the loop.

for number in integers:
    print(number)
```

```python
# For each iteration, irrespective of the value of the current item in the list (number), it will print the string 'yep!'
# Since there are five items in the list, 'yep!' will be printed five times. The output will look like this:

for number in integers:
    print('yep!')
```

```python
integers = [1,2,3,4,5]
```

```python
# The for loop is iterating over the list integers (which contains the numbers 1 through 5). For each iteration, 
# it takes the current element from the integers list, assigns it to the variable Jelly, and then executes the code 
# inside the loop.

The line print(Jelly + Jelly) adds the current number to itself (effectively multiplying it by 2), and then prints the result.

for Jelly in integers:
    print(Jelly + Jelly)
```

```python
# This code is creating a Python dictionary named ice_cream_dict. Dictionaries in Python are a type of data structure that 
# allow the storage of key-value pairs. In other words, each value in the dictionary has a unique key that can be used to 
# access it

ice_cream_dict = {'name': 'Alex Freberg', 'weekly intake': 5, 'favourite ice creams': ['MCC', 'Chocolate']}
```

```python
# The code you've given is a for loop that iterates over the values in the ice_cream_dict dictionary
# ice_cream_dict.values(): This is a method that returns a view object containing the values of the dictionary. 
# In this case, it will return 'Alex Freberg', 5, and ['MCC', 'Chocolate'], which are the values of ice_cream_dict.
# for cream in ice_cream_dict.values(): This for loop iterates over each value in the dictionary.
# print(cream): This line of code is executed for each iteration of the for loop. It prints the current value that 
# the loop is iterating over (which is referred to by the variable cream).

for cream in ice_cream_dict.values():
    print(cream)
```

```python
# this time it uses the items() method and unpacks both the key and value at the same time in the for loop.
# ice_cream_dict.items(): This method returns a view object that displays a list of a dictionary's key-value tuple pairs.
# the fact that we've used the terms 'key' and 'value makes no difference, they're replacable with other variables

for key, value in ice_cream_dict.items():
    print(key, '->', value)
```

## Nested For Loops

```python
# create two lists, 'flavours' and 'toppings'

flavours = ['Vanilla', 'Chocolate', 'Cookie Dough']
toppings = ['Hot Fudge', 'Oreos', 'Marshmallows']
```

```python
# This code creates all possible combinations of ice cream flavours ('Vanilla', 'Chocolate', 'Cookie Dough') 
# with toppings ('Hot Fudge', 'Oreos', 'Marshmallows') and prints each combination. It does this by using nested for loops: 
# the outer loop iterates over the flavours, and for each flavour, the inner loop iterates over all the toppings.

for one in flavours:
    for two in toppings:
        print(one, 'topped with', two)
```
