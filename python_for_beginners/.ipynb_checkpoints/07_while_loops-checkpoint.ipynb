# While Loops

```python
# This code uses a while loop to print numbers from 0 to 4. The while loop continues as long as the condition 
# number < 5 is true. In each iteration, it prints the current value of number and then increments number by 1
# remember that numbers start at 0, and if we didn't add the additional count at the bottom it would just loop forever

number = 0

while number < 5:
    print(number)
    number = number + 1
```

## Break Statements

```python
# the 'break' statement is used to exit or terminate a loop prematurely. When a 'break' statement is encountered within 
# a loop (for loop, while loop, or nested loops), it immediately stops the loop's execution and jumps to the next 
# statement after the loop.

number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
# including an 'else' statment, once the condition isn't met it will trigger the 'else' statement

number = 0

while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('No longer < 5')
```

```python
# if the 'break' trigger is met prior to the 'else' statement being triggered then the 'else' code won't execute

number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
else:
    print('No longer < 5')
```

## Continue Statement

```python
# The code uses a 'while' loop to increment the 'number' variable from 0 to 5. It prints the 'number' in each iteration 
# except when the 'number' is 3, where it skips that iteration. After the loop completes, it prints 'No longer < 5'

number = 0

while number < 5:
    number = number + 1
    if number == 3:
        continue
    print(number)
else:
    print('No longer < 5')
```
