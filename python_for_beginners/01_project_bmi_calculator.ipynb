# BMI Calculator

## BMI Calculation

#### BMI = (weight in pounds x 703) / (height in inches x height in inches)

## BMI Ranges

```python
# inputs
# weight in pounds, my example 178
# height in inches, my example 73

# outputs
# classification & health risk
```

```python
# creating the variables 'weight' & 'height', which allows the user to input their measurements
# we need to ensure the data type of these variables are integers so we define that in the logic

weight = int(input("Enter your weight in pounds: "))

height = int(input("Enter your height in inches: "))
```

```python
# incorporating the BMI calculation

weight = int(input("Enter your weight in pounds: "))

height = int(input("Enter your height in inches: "))

bmi = (weight * 703) / (height * height)

print(bmi)
```

```python
# creating the if logic to determine the bmi ranges

if bmi > 0:
    if (bmi < 18.5):
        print("You are Underweight.")
    elif (bmi <= 24.9):
        print("You are Normal Weight.")
    elif (bmi <= 29.9):
        print("You are Overweight.")        
    elif (bmi <= 34.9):
        print("You are Obsese.")        
    elif (bmi <= 39.9):
        print("You are Severely Obese.")        
    elif (bmi <= 39.9):
        print("You are Morbidly Obese.")           
    else:
        print("Enter valid inputs")
```

```python
# combining the syntax together to create a bmi_calc function

# This line defines a function called bmi_calc. This function doesn't take any arguments
def bmi_calc():
    
    # prompt for user to enter their name
    name = input("Enter your name: ")
    
    # prompt for user to enter their weight in pounds, the value is then converted to an integer 
    weight = int(input("Enter your weight in pounds: "))
    
    # prompt for user to enter their height in pounds, the value is then converted to an integer
    height = int(input("Enter your height in inches: "))
    
    # calculation for bmi based on variables weight & height
    bmi = (weight * 703) / (height * height)
    
    # starts a conditional block that will only execute if the calculated BMI is greater than 0
    if bmi > 0:
        # series of conditional checks that classify the user's BMI into various categories
        if (bmi < 18.5):
            print(name + ", you are Underweight.")
        elif (bmi <= 24.9):
            print(name + ", you are Normal Weight.")
        elif (bmi <= 29.9):
            print(name + ", you are Overweight.")        
        elif (bmi <= 34.9):
            print(name + ", you are Obsese.")        
        elif (bmi <= 39.9):
            print(name + ", you are Severely Obese.")        
        elif (bmi > 39.9):
            print(name + ", you are Morbidly Obese.")           
        # if the BMI is not greater than 0, this line will execute and print out "Enter valid inputs"
        else:
            print(name+ ", enter valid inputs")

# calls the function bmi_calc, which starts the process of prompting the user for their weight and height, 
# calculating their BMI, and printing out the classification of their BMI.
bmi_calc()
```

```python
# calling the saved function

bmi_calc()
```
