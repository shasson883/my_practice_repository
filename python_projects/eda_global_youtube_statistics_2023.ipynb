# EDA Global Youtube Statistics 2023

## 1. Understand the dataset

### Potential use cases

```python
"""

YouTube Analytics: Gain valuable insights into the success factors of top YouTube channels and understand what sets them 
apart from the rest.

Content Strategy: Discover the most popular categories and upload frequencies that resonate with audiences.

Regional Influencers: Identify influential YouTube creators from different countries and analyze their impact on a global scale.

Earnings Analysis: Explore the correlation between channel performance and estimated earnings.

Geospatial Visualization: Visualize the distribution of successful YouTube channels on a world map and uncover geographical 
trends.

Trending Topics: Investigate how certain categories gain popularity over time and correlate with world events.

"""
```

### Data attributes

```python
"""
rank: Position of the YouTube channel based on the number of subscribers
Youtuber: Name of the YouTube channel
subscribers: Number of subscribers to the channel
video views: Total views across all videos on the channel
category: Category or niche of the channel
Title: Title of the YouTube channel
uploads: Total number of videos uploaded on the channel
Country: Country where the YouTube channel originates
Abbreviation: Abbreviation of the country
channel_type: Type of the YouTube channel (e.g., individual, brand)
video_views_rank: Ranking of the channel based on total video views
country_rank: Ranking of the channel based on the number of subscribers within its country
channel_type_rank: Ranking of the channel based on its type (individual or brand)
video_views_for_the_last_30_days: Total video views in the last 30 days
lowest_monthly_earnings: Lowest estimated monthly earnings from the channel
highest_monthly_earnings: Highest estimated monthly earnings from the channel
lowest_yearly_earnings: Lowest estimated yearly earnings from the channel
highest_yearly_earnings: Highest estimated yearly earnings from the channel
subscribers_for_last_30_days: Number of new subscribers gained in the last 30 days
created_year: Year when the YouTube channel was created
created_month: Month when the YouTube channel was created
created_date: Exact date of the YouTube channel's creation
Gross tertiary education enrollment (%): Percentage of the population enrolled in tertiary education in the country
Population: Total population of the country
Unemployment rate: Unemployment rate in the country
Urban_population: Percentage of the population living in urban areas
Latitude: Latitude coordinate of the country's location
Longitude: Longitude coordinate of the country's location

"""
```

## Data Cleaning

### Check that we need all columns

```python
# in this df all the columns are useful so will be kept & not dropped
```

### Rename columns

```python
# replace '_' with ' ' and capitalize the first letter of each word
df.columns = df.columns.str.replace('_', ' ').str.title()

print(df.columns)
```

### Check for duplicated rows

```python
# uses the duplicated() method to mark duplicate rows

duplicates = df.duplicated()

# filters the DataFrame using the Boolean Series to examine the duplicates

duplicate_rows = df[duplicates]

# result is that no duplicate rows exist

print(duplicate_rows) 
```

### Dropping unwanted columns

```python
# dropping the rank column

df = df.drop(columns = 'Rank')
df
```

### Change data types

```python
# fills missing values in specified columns of the df with 0, and updates the df in place

df.fillna({
    'Video Views': 0,
    'Video Views Rank': 0,
    'Country Rank': 0,
    'Channel Type Rank': 0,
    'Video Views For The Last 30 Days': 0,
    'Subscribers For Last 30 Days': 0,
    'Created Year': 0,
    'Population': 0,
    'Urban Population': 0
}, inplace=True)


# converts the data types of specified columns in the DataFrame df to int64

df = df.astype({
    'Video Views': 'int64',
    'Video Views Rank': 'int64',
    'Country Rank': 'int64',
    'Channel Type Rank': 'int64',
    'Video Views For The Last 30 Days': 'int64',
    'Subscribers For Last 30 Days': 'int64',
    'Created Year': 'int64',
    'Population': 'int64',
    'Urban Population': 'int64'
})
```

```python
# displays concise information of the df

df.info()
```

### Identify missing values

```python
# counts the number of missing values (NaN) in each column of df

missing_values_count = df.isnull().sum()
missing_values_count[missing_values_count > 0]
```

```python
# replace missing values in categorical columns with "Unknown"

categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].fillna("Unknown")

# replace missing values in numerical columns with 0, excluding Latitude & Longitude columns

numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
numerical_columns = numerical_columns.difference(['Latitude', 'Longitude']) 
df[numerical_columns] = df[numerical_columns].fillna(0)

# check if there are any missing values remaining in the dataset
remaining_missing_values = df.isnull().sum().sum()
remaining_missing_values
```

### Clean & remove unwanted characters

```python
# matches any character that is NOT: a letter, a number, a whitespace character, a standard punctuation # mark (.,!?) a hyphen, 
# an apostrophe, and an ampersand

pattern = r'[^a-zA-Z0-9\s.,!?&\'-]'
```

```python
# replaces non-alphanumeric characters (excluding spaces, punctuation marks, and special characters listed) in the 'Youtuber' & 
# 'Title'columns in the df with an empty string

df['Youtuber'] = df['Youtuber'].str.replace('[^a-zA-Z0-9\s.,!?&\'-]', '')
df['Title'] = df['Title'].str.replace('[^a-zA-Z0-9\s.,!?&\'-]', '')

df
```

```python
# removing any whitespace or leading / trailing characters in 'Youtuber' column

df['Youtuber'] = df['Youtuber'].str.strip()
```

```python
# filters the 'Youtuber' column for rows which contain strings with unwanted characters
filtered_rows_youtuber = df[df['Youtuber'].str.contains(pattern, regex=True)]

# reviews the results
filtered_rows_youtuber
```

```python
# filters the 'Title' column for rows which contain strings with unwanted characters

filtered_rows_title = df[df['Title'].str.contains(pattern, regex=True)]

# reviews the results

filtered_rows_title
```

```python
# what is the average number of subscribers, views, and videos across all channels?

# average number of subscribers per channel

average_subscribers = int(round(df['Subscribers'].mean(), 0))
print("Average Number of Subscribers per Channel =", average_subscribers)

# average number of views per channel

average_views = int(round(df['Video Views'].mean(), 0))
print("Average Number of Views per Channel =", average_views)

# average number of videos per channel

average_videos = int(round(df['Uploads'].mean(), 0))
print("Average Number of Video Uploads per Channel =", average_videos)
```

```python
# what is the median number of subscribers, views, and videos

# median number of subscribers per channel

median_subscribers = int(round(df['Subscribers'].median(), 0))
print("Median Number of Subscribers Across Channels =", median_subscribers)

# median number of views per channel

median_views = int(round(df['Video Views'].median(), 0))
print("Median Number of Views Across Channels =", median_views)

# median number of videos per channel

median_videos = int(round(df['Uploads'].median(), 0))
print("Median Number of Videos Across Channels =", median_videos)
```

```python
# what is the minimum & maximum number of subscribers, views, and videos

# minimum number of subscribers per channel

min_subscribers = int(round(df['Subscribers'].min(), 0))
print("Minimum Number of Subscribers Across Channels = ", min_subscribers)

# minimum number of views per channel

min_views = int(round(df['Video Views'].min(), 2))
print("Minimum Number of Views Across Channels = ", min_views)

# minimum number of videos per channel

filtered_df_uploads = df[df['Uploads'] > 0]
min_videos = int(round(filtered_df_uploads['Uploads'].min(), 2))
print("Minimum Number of Videos Across Channels = ", min_videos)

# maximum number of subscribers per channel

max_subscribers = int(round(df['Subscribers'].max(), 0))
print("Maximum Number of Subscribers Across Channels = ", max_subscribers)

# maximum number of views per channel

max_views = int(round(df['Video Views'].max(), 2))
print("Maximum Number of Views Across Channels = ", max_views)

# maximum number of videos per channel

max_videos = int(round(df['Uploads'].max(), 2))
print("Maximum Number of Videos Across Channels = ", max_videos)
```

```python
# who are top YouTubers?

# top 10 YouTube channels

selected_columns = ['Youtuber', 'Subscribers']
top_10_youtubers = df.loc[0:9, selected_columns]

# sorting values so that the YouTubers with the highest number of subscribers are at the top
top_10_youtubers_sorted = top_10_youtubers.sort_values(by = 'Subscribers', ascending = True)

# plotting as a horizontal bar chat
top_10_youtubers_sorted.plot.barh(x = 'Youtuber', y = 'Subscribers', stacked = True, color = 'red')

plt.xlabel('# Subscribers')
plt.ylabel('YouTuber')
plt.title('Top 10 YouTube Channels by No. Subscribers (2023)')
plt.show()
```

```python
# what are the number of YouTubers per category

# setting the size of the plot
plt.figure(figsize = (15, 8))

# creating a count plot using Seaborn
sns.countplot(data = df, x = 'Category', order = df['Category'].value_counts().index, color = 'red')

# adjusting the title & labels of the chart
plt.title('No. of YouTube Channels by Category (2023)', fontsize = 18)
plt.xlabel('Category', fontsize = 15)
plt.ylabel('# Channels', fontsize = 15)
plt.xticks(rotation = 45, ha = 'right')

# displaying the chart output
plt.tight_layout()
plt.grid(False)

plt.show()
```

```python
# what are the top 10 YouTube categories by number of channels?

# calculate the number of YouTubers in each category
category_counts = df['Category'].value_counts()

# select the top 10 categories based on the number of YouTubers
top_10_categories = category_counts.head(10)

# sorting values so that the YouTubers with the highest number of subscribers are at the top
top_10_categories_sorted = top_10_categories.sort_values(ascending = True)

# plotting as a horizontal bar chat
top_10_categories_sorted.plot.barh(stacked = True, color = 'red')

# adjusting the title & labels of the chart
plt.xlabel('# Channels')
plt.ylabel('Category')
plt.title('Top 10 YouTube Channels Categories (2023)') 

# displaying the chart output
plt.show()
```

```python
# what categories have the most views on average?

# calculating the average no. video views per category
category_average_views = df.groupby('Category')['Video Views'].mean() / 1e9

# sorting values so that the categories with the highest no. of video views are at the top
category_average_views = category_average_views.sort_values(ascending = True)

# plotting as a horizontal bar chat
category_average_views.plot.barh(stacked = True, color = 'red')

# adjusting the title & labels of the chart
plt.xlabel('# Average Video Views (bn.)')
plt.ylabel('Category')
plt.title('Average No. of Video Views by YouTube Channel Category (2023)') 

# displaying the chart output
plt.show()
```

```python
# which categories have the most subscribers on average?

# calculating the average no. subscribers per category reflected in millions
category_average_subscribers = df.groupby('Category')['Subscribers'].mean() / 1e6
# print(category_average_subscribers)

# sorting values so that the categories with the highest average no. of subscribers are at the top
category_average_subscribers = category_average_subscribers.sort_values(ascending = True)

# plotting as a horizontal bar chat
category_average_subscribers.plot.barh(stacked = True, color = 'red')

# adjusting the title & labels of the chart
plt.xlabel('# Average Subscribers (mn.)')
plt.ylabel('Category')
plt.title('Average No. of Subscribers by YouTube Channel Category (2023)') 

# displaying the chart output
plt.show()
```
